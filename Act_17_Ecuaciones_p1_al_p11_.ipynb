Problema 1: $y - 5 = 12$
* Planteamiento: A un número $y$ se le restan 5 y da 12.
* Despeje: El 5 que está restando pasa al otro lado sumando.
* Solución: $y = 12 + 5$ → $y = 17$

Problema 3: $12 = m + 3m$
* Planteamiento: La suma de un número más su triple es igual a 12.
* Simplificación: Agrupamos las "m": $1m + 3m = 4m$.
* Entonces: $12 = 4m$.
* Despeje: El 4 que multiplica a $m$ pasa dividiendo al 12.
+Solución: $m = \frac{12}{4}$ → $m = 3$

Problema 5: $5b + 2b - 3b = 16$
* Planteamiento: Operaciones combinadas con una misma incógnita.
* Simplificación: Resolvemos de izquierda a derecha: $5b + 2b = 7b$, luego $7b - 3b = 4b$.

* Ecuación resultante: $4b = 16$.
* Despeje: El 4 pasa dividiendo.
* Solución: $b = \frac{16}{4}$ → $b = 4$

 Problema 7: $7y - 5 = 3y + 7$
* Planteamiento: Incógnitas en ambos lados de la igualdad.
* Agrupación: Pasamos $3y$ a la izquierda (restando) y el 5 a la derecha (sumando).
 * $7y - 3y = 7 + 5$
 * $4y = 12$
 * Solución: $y = \frac{12}{4}$ → $y = 3$

Problema 9: $8n - 6 = 4n + 10$
* Agrupación: Pasamos $4n$ a la izquierda y el 6 a la derecha.
* $8n - 4n = 10 + 6$
* $4n = 16$
* Solución: $n = \frac{16}{4}$ → $n = 4$

Problema 11: $10b - 8 = 4b + 10$
* Agrupación: Pasamos $4b$ a la izquierda y el 8 a la derecha.
* $10b - 4b = 10 + 8$
* $6b = 18$
* Solución: $b = \frac{18}{6}$ → $b = 3$

```python
# Problema 1: y - 5 = 12
y = 12 + 5
print(f"Solución Problema 1 (y): {y}")

# Problema 3: 12 = 4m
m = 12 / 4
print(f"Solución Problema 3 (m): {int(m)}")

# Problema 5: 4b = 16
b = 16 / 4
print(f"Solución Problema 5 (b): {int(b)}")

# Problema 7: 7y - 3y = 7 + 5 -> 4y = 12
y_7 = (7 + 5) / (7 - 3)
print(f"Solución Problema 7 (y): {int(y_7)}")

# Problema 9: 8n - 4n = 10 + 6 -> 4n = 16
n_9 = (10 + 6) / (8 - 4)
print(f"Solución Problema 9 (n): {int(n_9)}")

# Problema 11: 10b - 4b = 10 + 8 -> 6b = 18
b_11 = (10 + 8) / (10 - 4)
print(f"Solución Problema 11 (b): {int(b_11)}")
```
