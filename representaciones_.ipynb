## problema 1
La suma de un número más 9 es igual a 21Paso 1: Análisis y ResoluciónPara resolver esto, debemos seguir un orden lógico de traducción:Planteamiento: Identificamos la incógnita. "Un número" será nuestra variable $x$. La frase "la suma de un número más 9" se escribe como $x + 9$. "Es igual a 21" completa la ecuación.Ecuación: $x + 9 = 21$Despeje de la variable: Para dejar a la $x$ sola, debemos mover el 9 al otro lado del signo igual. Como el 9 está sumando, pasa haciendo la operación contraria (restando).$x = 21 - 9$Solución: Realizamos la operación aritmética.
$x = 12$

```python
# Problema 1: Hallar un número x tal que x + 9 = 21
resultado_total = 21
numero_conocido = 9

# Despejamos x: x = 21 - 9
x = resultado_total - numero_conocido

print(f"El número buscado es: {x}")
```

#### Problema 2
Las edades de Pedro y AnaPaso 1: Análisis y ResoluciónEste problema requiere relacionar dos variables basándonos en una referencia común.Planteamiento:Edad de Ana: La definimos como
$x$ (ya que la edad de Pedro depende de la de ella).Edad de Pedro: "El doble de edad que su hermana", es decir, $2x$.La suma de ambos es 36: $x + 2x = 36$.Reducción y Despeje:Sumamos las términos semejantes: $3x = 36$.Para despejar $x$, el 3 que está multiplicando pasa al otro lado dividiendo.$x = \frac{36}{3}$Solución:$x = 12$ (Edad de Ana).$2x = 2 \cdot 12 = 24$ (Edad de Pedro).Comprobación: $12 + 24 = 36$. ¡Correcto!

```python
# Problema 2: Edades de Pedro (2x) y Ana (x)
suma_total = 36

# La ecuación es 3x = 36, por lo tanto x = 36 / 3
edad_ana = suma_total / 3
edad_pedro = edad_ana * 2

print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
